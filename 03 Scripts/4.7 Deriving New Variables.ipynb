# 03 User-defined function

```python
# Defining a new function - price_label
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif (row['prices'] > 15):
        return 'High range'
    else: 
        return 'Not enough data'
```

```python
# Test the new function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check if the function worked
df['price_range'].value_counts(dropna = False)
```

# 04 Task

```python
# Add columns for new label on busy days per client request
# Add comment regarding busy days
result = []

for value in ords_prods_merge["order_day_of_week"]:
  if value in [0,1]:
    result.append("Busiest days")
  elif value in [3,4]:
    result.append("Slowest days")
  else:
    result.append("Regularly busy")
```

```python
ords_prods_merge['busiest_days_new'] = result
```

```python
ords_prods_merge['busiest_days_new'].value_counts(dropna=False)
```

```python
ords_prods_merge[['order_day_of_week', 'busiest_days_new']].head(60)
```

#### It seems that the busiest days/slowest days label are correct based on the first 60 rows

```python
# Now check the busiest hour of the day
# Check frequency of orders per hour
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

#### Busiest hours of the day is from 10:00 to 15:00, while the slowest is from 02:00-05:00.

```python
# Add columns for new label on busy hours of day
# Add comment regarding busy hours
result_hour = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value in [10,11,12,13,14,15]:
    result_hour.append("Most orders")
  elif value in [2,3,4,5]:
    result_hour.append("Fewest orders")
  else:
    result_hour.append("Average orders")
```

```python
ords_prods_merge['busiest_period_of_day'] = result_hour
```

```python
# Frequency of periods
ords_prods_merge['busiest_period_of_day'].value_counts(dropna=False)
```
