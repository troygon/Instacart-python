#### Looking at both columns, we can see that we can use the product_id as the key in combining them. Since the information on the df_prods is the details of the information of the df_orders_products_combined, we can just join df_prods to the df_orders_products_combined through inner join via the product_id

```python
# Test merge without overwriting
pd.merge(df_orders_products_combined,df_prods, on = ['product_id'], indicator = 'merge_2')
```

```python
# Merging the dataframe
df_merged_ords_prods = df_orders_products_combined.merge(df_prods, on = 'product_id', indicator = 'merge_2')
```

```python
df_merged_ords_prods.head()
```

```python
# Checking how the data is merged
df_merged_ords_prods['merge_2'].value_counts()
```

```python
# We can drop the orginal _merge column in this dataframe since it will be confusing and it is not relevant in this dataframe anymore
df_merged_ords_prods = df_merged_ords_prods.drop(columns='_merge')
```
