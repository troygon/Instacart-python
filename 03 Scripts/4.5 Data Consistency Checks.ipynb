```python
# Create dataframe with NaN values
blank_days = df_ords_new[df_ords_new['days_from_last_order'].isnull() == True]
blank_days
```

 ## Suspected case where the NaN values are there is because this is the first instance that the user_id is ordering, therefore no previous order. Will check the blank_days dataframe if we have order_number not equal to one to confirm if this is the case

```python
# Filter rows in 'blank_days' where 'order_number' is not equal to 1
filtered_blank_days = blank_days[blank_days['order_number'] != '1']
filtered_blank_days
```

## Confirming all NaN values are for first time orders. We will retain the NaN values to indicate this is their first order, but will add another column to clarify this is their first time ordering

```python
# Add new_customer column for NaN values in days_from_last_order
df_ords_clean = df_ords_new
df_ords_clean['new_customer'] = df_ords_new['days_from_last_order'].isnull() == True
df_ords_clean
```

## Successfully added column new_customer with boolean values
